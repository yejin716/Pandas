조건(conditon)을 걸어, 데이터 프레임에서 특정 데이터 값 추출

```python
#CRIM < 1 이면서 target 24.0이상인 데이터 추출

#1단계
df.loc[(df['CRIM'] < 1) & (df['target'] >= 24)]
```

```python
#2단계
df.loc[(df['CRIM'] < 1) & (df['target'] >= 24), cols]
```

```python
#3단계
df.loc[(df['CRIM'] < 1) & (df['target'] >= 24), cols + ['target']]
```

```python
#CRIM < 0.8 이면서 target 20.0이상인 데이터 추출

cols = ['CRIM', 'ZN', 'CHAS']
condition =(df['CRIM'] < 0.8) & (df['target'] >= 20)
df.loc[condition, cols + ['target']]
```

```python
def 부동산가격결정요인():
    cols = ['CRIM', 'ZN', 'CHAS']
    condition =(df['CRIM'] < 0.8) & (df['target'] >= 20)
    
    result = df.loc[condition, cols + ['target']]
    return result
```
