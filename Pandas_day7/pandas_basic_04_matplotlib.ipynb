```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [1,2,3,4,5]

plt.plot(x,y)
plt.show() #그림만 나오게
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]

plt.plot(x,y)
plt.title("First Plot")
plt.xlabel("x")
plt.xlabel("y")
plt.show()
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]
fig, ax = plt.subplots()
ax.plot(x,y)
ax.set_title("First Plot")
ax.set_xlabel("x")
ax.set_ylabel("y")

fig.set_dpi(300) #해상도 (dot per inch)
fig.savefig("first_plot.png") #그래프를 사진으로 저장 
plt.show()
```

```python
#여러 개 그래프 그리기 

x = np.linspace(0, np.pi*4, 100) #pi*4를 100등분 
fig, axes = plt.subplots(2,1)
axes[0].plot(x, np.sin(x))
axes[1].plot(x, np.cos(x))
plt.show()
```

Matplotlib 주요 그래프 

```python
#선 그래프 (line plot)
fig, ax = plt.subplots()
x = np.arange(15)
y = x ** 2

ax.plot(x,y,color="r")
ax.plot(x+2,y,color="b")

plt.show()
```

```python
#line style

x = np.arange(10)
fig, ax = plt.subplots()
ax.plot(x, x, linestyle="-")
# solid
ax.plot(x, x+2, linestyle="--")
# dashed
ax.plot(x, x+4, linestyle="-.")
# dashdot
ax.plot(x, x+6, linestyle=":")
#dotted
```

```python
#color

x = np.arange(10)
fig, ax = plt.subplots()
ax.plot(x, x, color="r")
ax.plot(x, x+2, color="green")
ax.plot(x, x+4, color='0.8') #그레이스케일(흑백) 0~1사이로 나타냄 0:검은색, 1:흰색
ax.plot(x, x+6, color="#524FA1")
```

```python
#marker 
x = np.arange(10)
fig, ax = plt.subplots()
ax.plot(x, x, marker=".")
ax.plot(x, x+2, marker="o")
ax.plot(x, x+4, marker='v')
ax.plot(x, x+6, marker="s")
ax.plot(x, x+8, marker="*")
```

```python
#축 경계 조정하기 
x = np.linspace(0, 10, 100)
fig, ax = plt.subplots()
ax.plot(x, np.sin(x))
ax.set_xlim(-2,12)
ax.set_ylim(-1.5, 1.5)
```

```python
#범례

fig,ax = plt.subplots()
ax.plot(x, x, label='y=x')
ax.plot(x, x**2, label='y=x^2')

ax.set_title("legend")
ax.set_xlabel("x")
ax.set_ylabel("y")

ax.legend(loc='upper right',
        shadow=True,
        fancybox = True,
        borderpad=2)
plt.show()
```

```python
#scatter 
fig, ax = plt.subplots()
x = np.arange(10)
ax.plot(
x, x**2, "o",
markersize=15,
markerfacecolor='white',
markeredgecolor="blue")
```

```python
#bubble 형태 scatter 

fig, ax = plt.subplots()
x = np.random.randn(50) 
y = np.random.randn(50)
colors = np.random.randint(0, 100, 
50)
sizes = 500 * np.pi * np.random.rand(50) ** 2
ax.scatter(
x, y, c=colors, s=sizes, alpha=0.3) #투명도 alpha 0~1사이 
```

```python
#bar plot 

x = np.arange(10)
fig, ax = plt.subplots(figsize = (8,4))
ax.bar(x,x*2)
```

```python
#누적 bar그래프 

x = np.random.rand(3)
y = np.random.rand(3)
z = np.random.rand(3)
data = [x,y,z]

fig, ax = plt.subplots()
x_ax = np.arange(3)

for i in x_ax:
    ax.bar(x_ax, data[i],
    bottom=np.sum(data[:i], axis=0))
ax.set_xticks(x_ax)
ax.set_xticklabels(["A","B","C"])
```

```python
#histograme
fig, ax = plt.subplots()
data = np.random.randn(1000)
ax.hist(data, bins=50)
```
