```python
#표준화 (평균 0, 분산 1)
#각 데이터 값이 평균을 기준으로 얼마나 떨어져 있냐 (편차)
#평균(0), 편차 >> 표준편차(1)
#z-score : 각 데이터 값에서 평균 뺀 후 (x-x_bar) 표준편차(sigma) 로 나누어 준 값 
#
```

```python
#정규화 
#데이터의 범위를 0 - 1로 변환하여 데이터 분포를 조정하는 것 
#(x-최소값) / (최대값-최소값)
#(최대값-최소값) : 데이터가 존재하는 범위 
#min-max scaler()
```

```python
fin_min = np.min(df['기말'])
fin_max = np.max(df['기말'])
print(fin_min)
print(fin_max)
```

```python
df['기말_정규화'] = (df['기말'] - fin_min) / (fin_max-fin_min)
```
